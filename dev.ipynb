```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://www.semafor.com/'
soup = BeautifulSoup(requests.get(url).content, 'lxml')
soup
```

```python
set(sorted([a['href'] for a in soup.find('div', {'class': re.compile('styles_grid')}).find_all('a', {'href': re.compile("/article/")})]))
```
